# Exponential Smoothing in Python 

First we will import the necessary libraries, then we will pass the csv into a dataframe using pandas. Then we use the 'df.head()' to get an idea of what the csv looks like. The provides monthly totals of a US airline passengers from 1949 to 1960. Our data set can be found [here](https://www.kaggle.com/chirag19/air-passengers).

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

df = pd.read_csv('AirPassengers.csv',index_col='Month',parse_dates=True)
df.index.freq = 'MS'
df.head()
```

Now we need to separate the data into training and testing. Then we will plot it to see the level, trend, and seasonality of the data.

```python
train_data = df.iloc[:108]
test_data = df.iloc[108:]

train_data['#Passengers'].plot(legend=True,label='TRAIN')
test_data['#Passengers'].plot(legend=True,label='TEST',figsize=(12,8));
```

We can clearly see the the trend is upward with a seasonality of approximately 12 months. We can use the 'seasonal_decompose' module to get the trend and seasonality of the data. 

```python
from statsmodels.tsa.seasonal import seasonal_decompose
seasonal_decompose(train_data['#Passengers']).plot();
```

## Simple Exponential Smoothing

Now we will look at the simple exponential smoothing model using this data. Note that alpha was found using a model linked [here](https://pandas.pydata.org/pandas-docs/stable/user_guide/computation.html#exponentially-weighted-windows).

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing
span = 12
alpha = 2/(span+1)
model = SimpleExpSmoothing(train_data['#Passengers']).fit(smoothing_level=alpha)
test_predictions = model.forecast(36).rename('SES Forecast')
```

We will now plot to see the our prediction. 

```python
train_data['#Passengers'].plot(legend=True,label='TRAIN')
test_data['#Passengers'].plot(legend=True,label='TEST',figsize=(12,8))
test_predictions.plot(legend=True,label='PREDICTION');
```

As you can see, this type of smoothing is not effective for this particular dataset, espcially due to the data containing a trend and seasonality. We will take a look at the mean squared error to get an idea of accurate we are.

```python
from sklearn.metrics import mean_squared_error
rms_error = np.sqrt(mean_squared_error(test_data['#Passengers'],test_predictions))
print("RMS Error is ",rms_error)
```

This also shows how unaccurate our prediction is seeing how far the RMS Error is from 1. 

## Double Exponential Smoothing

Double exponential smoothing involves alpha to smooth the level, but what if there is a trend, thats where this model comes in. There are 2 types of trends, this includes Additive Trends & Multiplicative Trends. The difference between the two is if the trend is linear or exponential respectively. The introduction fo the beta parameter is used to smooth the trend. In the next example, the model will use default values for alpha and beta.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
double_model = ExponentialSmoothing(train_data['#Passengers'],trend='add').fit()
test_predictions = double_model.forecast(36).rename('DES Forecast')
```

Now we will plot it.

```python
train_data['#Passengers'].plot(legend=True,label='TRAIN')
test_data['#Passengers'].plot(legend=True,label='TEST',figsize=(12,8))
test_predictions.plot(legend=True,label='PREDICTION');
```

Now our prediction is going in the right direction, but is still missing a critical feature (seasonality), therefore out prediction is still very off. We will take a look at this graphs RMS Error now to compare it to our previous error. 

```python
rms_error = np.sqrt(mean_squared_error(test_data['#Passengers'],test_predictions))
print("RMS Error for DES is ",rms_error)
```

We can see that the error has gone down, but not by much. This is still a very high number, indicating our prediction is still not optimized to its fullest potential.

## Triple Exponential Smoothing

Just like Double Exponential Smoothing, this model will contain alpha and beta, but there is an introduction of a new variable called gamma. This expectedly smooths the seasonality. However unlike the last two models, this model requires a period. We established earlier in this document that the period is 12 months, due in part to the data being monthly over the course of 11 years.

```python
from statsmodels.tsa.holtwinters import ExponentialSmoothing
triple_model = ExponentialSmoothing(train_data['#Passengers'],trend='add',seasonal='add',seasonal_periods=12).fit()
test_predictions = triple_model.forecast(36).rename('TES Forecast')
```

Now we will plot it.

```python
train_data['#Passengers'].plot(legend=True,label='TRAIN')
test_data['#Passengers'].plot(legend=True,label='TEST',figsize=(12,8))
test_predictions.plot(legend=True,label='PREDICTION');
```

With all of our components finally being used, we can finally get a more accurate representaion of the forecasted data. This model, whether additive or exponential, is most commonly used among the exponential smoothing model family. Lastly we will take a look at the RMS Error to see how much more accurate we were this time.

```python
rms_error = np.sqrt(mean_squared_error(test_data['#Passengers'],test_predictions))
print("RMS Error for TES is ",rms_error)
```

Our RMS Error is approx. 30. This number is much closer to one than our other RMS Errors, which were approx. 87 and 103, indicating this model to be the most effecient for this dataset.
